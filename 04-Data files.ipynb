2.1 На основе файла contributors_sample.json создайте словарь следующего вида:

{
    должность: [список username людей, занимавших эту должность]
}

```python
import json

with open('contributors_sample.json') as f:
    data = json.load(f)

    positions = {}

    for contributor in data:
        for position in contributor['jobs']:
            if position in positions:
                positions[position].append(contributor['username'])
            else:
                positions[position] = [contributor['username']]

print(positions)
```

2.2 Сохраните результаты в файл job_people.pickle и в файл job_people.json с использованием форматов pickle и JSON соответственно. Сравните объемы получившихся файлов. При сохранении в JSON укажите аргумент indent.

```python
import pickle
import json

with open('contributors_sample.json') as f:
    data = json.load(f)

    positions = {}

    for contributor in data:
        for position in contributor['jobs']:
            if position in positions:
                positions[position].append(contributor['username'])
            else:
                positions[position] = [contributor['username']]

print(positions)

with open('job_people.pickle', 'wb') as f:
    pickle.dump(positions, f)

with open('job_people.json', 'w') as f:
    json.dump(positions, f, indent=4)

import os

print(f"pickle file size: {os.path.getsize('job_people.pickle')} bytes")
print(f"JSON file size: {os.path.getsize('job_people.json')} bytes")
```

# XML

3.1 По данным файла steps_sample.xml сформируйте словарь с шагами по каждому рецепту вида {id_рецепта: ["шаг1", "шаг2"]}. Сохраните этот словарь в файл steps_sample.json

```python
import xml.etree.ElementTree as ET
import json

tree = ET.parse('steps_sample.xml')
root = tree.getroot()

steps_dict = {}

for recipe in root:
    recipe_id = recipe.attrib['id']
    steps = []
    for step in recipe.findall('./step'):
        steps.append(step.text)
    steps_dict[recipe_id] = steps

with open('steps_sample.json', 'w') as f:
    json.dump(steps_dict, f)
```

3.2 По данным файла steps_sample.xml сформируйте словарь следующего вида: кол-во_шагов_в_рецепте: [список_id_рецептов]

```python
import xml.etree.ElementTree as ET

tree = ET.parse('steps_sample.xml')
root = tree.getroot()

recipe_steps_dict = {}
for recipe in root.findall('./recipe'):
    recipe_id = recipe.find('id').text
    steps_count = len(recipe.findall('./cook_step'))
    if steps_count in recipe_steps_dict:
        recipe_steps_dict[steps_count].append(recipe_id)
    else:
        recipe_steps_dict[steps_count] = [recipe_id]

print(recipe_steps_dict)
```

3.4 Загрузите данные из файла recipes_sample.csv (ЛР2) в таблицу recipes. Для строк, которые содержат пропуски в столбце n_steps, заполните этот столбец на основе файла steps_sample.xml. Строки, в которых столбец n_steps заполнен, оставьте без изменений.

```python
import pandas as pd
from xml.etree import ElementTree

recipes = pd.read_csv("recipes_sample .csv")

# заполнение пропусков в столбце n_steps на основе файла steps_sample.xml
tree = ElementTree.parse("steps_sample.xml")
root = tree.getroot()

for recipe in root.findall("recipe"):
    recipe_id = recipe.get("id")
    n_steps = len(recipe.findall("step"))
    recipes.loc[recipes.id == recipe_id, "n_steps"] = n_steps

print(recipes)
```
